```python
def lspl(fid):
    import numpy as np
    import matplotlib.pyplot as plt
    plt.switch_backend('agg')
    sample_interval=200
    max_interval=200*1000
    #fid=50237
    f = open("RDB_run.sh.o%d"%(fid),"r")
    lines = f.readlines() 
    train_loss=np.zeros((max_interval,1))
    val_loss=np.zeros((max_interval,1))
    for line in lines:
        tempp=line.split(" ")
        if tempp[0].isdigit():
            train_loss[int(tempp[0])]=float(tempp[-1])
            pre_ind=int(tempp[0])
            #print(float(tempp[0]))
        temp=line.split(":")
        if temp[0]=='Loss':
            #print(tempp[0])
            val_loss[pre_ind]=float(temp[-1])
    
    return train_loss,val_loss
            
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.switch_backend('agg')
fid=74851
train_loss,val_loss=lspl(fid)
ind=np.linspace(0,199999,200000)
plt.plot(ind,train_loss,'r',ind[0:200000:200],val_loss[0:200000:200])
plt.title('Loss function of DN Degradation mode with scale 3')
plt.xlabel('Iteration')
plt.legend(['train_MAE','val_MAE'])
plt.savefig('loss%d.png'%(fid),dpi=1200)
```
